## Image Classification (Alpacas vs Llamas)

The objective of this project is to train a neural network for an image classification problem. The model is aimed to differentiate between alpacas and llamas.

    1. Getting image data: Use MS Azure Bing Image Search to obtain image search results.
    2. Get the data in the DataLoader format - splitting them into train, valid and test sets.
    3. Data Augmentation: apply some variations to the input data.
    4. Train a neural network using PyTorch cnn_learner method.
    5. Evaluate the model.
    
    This project uses PyTorch, fastai and fastbook libraries.

Obtain image search results for aplacas and llamas. The free tier MS Azure account will can provide 150 images per keyword search.

```python
animals_types = 'alpaca', 'llama'
path = Path('animals')

if not path.exists():
    path.mkdir()
    
    for o in tqdm(animals_types):
        print('Animal type: ', o)
        dest = (path/o)
        dest.mkdir(exist_ok=True)
        
        results = search_images_bing(key, f'{o}')
        download_images(dest, urls=results.attrgot('contentUrl'))
        sleep(1)
```

Pass the images to DataBlock, splitting the dataset into training and validation sets.

```python
animals = DataBlock(
            blocks=(ImageBlock, CategoryBlock),
            get_items=get_image_files,
            splitter=RandomSplitter(valid_pct=0.2),
            get_y=parent_label,
            item_tfms=Resize(128))
```

```python
dls = animals.dataloaders(path, num_workers=0)
```

Viewing a snapshot of the data.

```python
dls.valid.show_batch(max_n=24, nrows=3)
```

```python
animals = animals.new(
    item_tfms=RandomResizedCrop(224, min_scale=0.5),
    batch_tfms=aug_transforms())

dls = animals.dataloaders(path, num_workers=0)
```

```python
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(5)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```
